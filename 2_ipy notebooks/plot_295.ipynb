```python
import plotly.graph_objects as go

Algorithms = ['CBV', 'ALS']
# Time = [16025, 1642000 ]
Time = [16.025, 1642]

fig = go.Figure(data=[go.Bar(
            x=Algorithms, y=Time,
            text=Time,
            textposition='auto',
        )])
fig.update_layout(title_text='Runtime Comparision for 100k movielens',
    yaxis_title="Runtime in seconds")
fig.update_traces(marker_color='#a284e0')
```

```python
import plotly.graph_objects as px

x = ['CBV', 'ALS']
y5 = [0.135563, 0.056790]
y10 = [0.119366, 0.053439]
y25 = [0.098662,0.057000]
y50 = [0.079930,0.053679]
y100 = [0.061866,0.047839]
plotp = px.Figure(data=[go.Bar(
    name = '5 Recommendations',
    x = x,
    y = y5,
    text = y5,
    textposition='auto',
    marker_color='yellow'
   ),
    go.Bar(
    name = '10 Recommendations',
    x = x,
    y = y10,
    text = y10,
    textposition='auto'
    , marker_color='green'),
    go.Bar(
    name = '25 Recommendations',
    x = x,
    y = y25
    , marker_color='#808000',
    text = y25,
    textposition='auto'
   ),
                       go.Bar(
    name = '50 Recommendations',
    x = x,
    y = y50
    , marker_color='#004d00',
    text = y50,
    textposition='auto'
   ),
                       go.Bar(
    name = '100 Recommendations',
    x = x,
    y = y100
    , marker_color=' #003400',
    text = y100,
    textposition='auto'
   )
])
                   
plotp.update_layout(title_text='Precision',
    yaxis_title="Precision score",barmode='group')
plotp.show()
```
